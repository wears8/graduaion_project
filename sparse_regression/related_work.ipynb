## 一、考虑到在进行稀疏回归时，对数据的处理问题，故而探究一下需要如何处理数据

### 1.数据标准化，采用正态分布

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('../improved_data/xy_orbit.csv')
array = df.to_numpy()
array = array.T
t = array[0]
yy = array[1:]

plt.figure(figsize=(8, 6))
plt.plot(yy[0], yy[1], label='Trajectory')
plt.scatter(yy[0,0], yy[1,0], color='red', label='Start')
plt.scatter(yy[0,-1], yy[1,-1], color='g', label='Start')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trajectory of the System')
plt.legend()
plt.grid()
plt.axis('equal')  # 确保 x 和 y 轴比例相同
plt.show()

scaler = StandardScaler()
yy_scaled = scaler.fit_transform(yy)

plt.figure(figsize=(8, 6))
plt.plot(yy_scaled[0], yy_scaled[1], label='Trajectory')
plt.scatter(yy_scaled[0,0], yy_scaled[1,0], color='red', label='Start')
plt.scatter(yy_scaled[0,-1], yy_scaled[1,-1], color='g', label='Start')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trajectory of the System')
plt.legend()
plt.grid()
plt.axis('equal')  # 确保 x 和 y 轴比例相同
plt.show()

```

#### 很明显，这样的标准化，尽管将尺度缩小了，但却改变了原有的分布，考虑到这是数据是对应时间节点的，这样的处理会改变系统的原有的运动。
